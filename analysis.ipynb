```python
# Import the relevant libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style('whitegrid')
```

```python
# Load the raw data
raw_data = pd.read_csv('customer_acquisition_data.csv')
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.describe(include='all')
```

```python
# Finding unique channels used for sale in dataset
raw_data['channel'].unique()
```

```python
# Plotting the graph between channel and cost
raw_data.groupby('channel')['cost'].sum().plot.bar()
plt.ylabel('Cost')
plt.title('Customer Acquisition cost by Channel')
```

According to the plot abobe we can say that paid advertisement is the most expensive channel whereas email marketing is the least expensive channel in terms of cost.

```python
# Plotting the graph between channel and conversion rate
raw_data.groupby('channel')['conversion_rate'].sum().plot.bar()
plt.ylabel('Conversion Rate')
plt.title('Conversion rate by Channel')
```

Social media is the most effective channel for converting customers, while paid advertising is the least effective.

```python
# Plotting the pie chart between channel and revenue
raw_data.groupby('channel')['revenue'].sum().plot.pie(autopct='%1.1f%%')
plt.title('Total revenue by Channel')
```

According to the plot above we can say that email marketing is the most profitable channel of marketing. But there is not huge difference between revenue generatin from all the other channels to call any channel the least profitable channel

```python
# Calculating return on investment
raw_data['roi'] = raw_data['revenue']/raw_data['cost']
```

```python
# Plotting the graph between channel and return on investment
raw_data.groupby('channel')['roi'].sum().plot.bar()
plt.ylabel('Return on Investment')
plt.title("Return on investment by channel")
```

ROI from email marketing is way higher than all the other channels while ROI from paid advertisements is the lowest.

```python
# Calculating the customer life time value
raw_data['cltv'] = (raw_data['revenue']-raw_data['cost'])*raw_data['conversion_rate']/raw_data['cost']
```

```python
# Plotting the graph between channel and customer life time value
raw_data.groupby('channel')['cltv'].sum().plot.bar()
plt.ylabel('Customer Lifetime value')
plt.title('Customer Lifetime values by channel')
```

The customer lifetime value from Social Media and the referral channels is the highest.
